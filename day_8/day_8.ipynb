````python
def find(parents, i):
    if parents[i] == i:
        return i

    parents[i] = find(parents, parents[i])
    return parents[i]
````

````python
def union(parents, i, j):
    pi = find(parents, i)
    pj = find(parents, j)

    parents[pj] = pi
````

````python
from collections import defaultdict

sizes = defaultdict(int)
for node in parents.values():
    root = find(parents, node)
    sizes[root] += 1

ans = 1
for s in sorted(sizes.values())[-3:]:
    ans *= s
````
